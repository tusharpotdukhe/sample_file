```python
import json
from collections import Counter
```

```python
def find_duplicate_labels(json_file):
    with open(json_file, 'r', encoding='utf-8') as file:
        data = json.load(file)
    
    labels = []
    
    # Extract labels (assuming labels are keys or values in a dictionary/list structure)
    def extract_labels(obj):
        if isinstance(obj, dict):
            for key, value in obj.items():
                labels.append(key)
                extract_labels(value)
        elif isinstance(obj, list):
            for item in obj:
                extract_labels(item)
        elif isinstance(obj, str):
            labels.append(obj)
    
    extract_labels(data)
    
    # Count occurrences of each label
    label_counts = Counter(labels)
    
    # Find duplicates
    duplicates = {label: count for label, count in label_counts.items() if count > 1}
    
    return duplicates

if __name__ == "__main__":
    json_file = r"C:\Users\tushar.potdukhe\Desktop\TW data\duplicates_labels\duplicates_labels\scooter\CPG62D1Y_460_6033_2_187.json"  # Change this to your JSON file
    duplicates = find_duplicate_labels(json_file)
    
    if duplicates:
        print("Duplicate labels found:")
        for label, count in duplicates.items():
            print(f"{label}: {count} times")
    else:
        print("No duplicate labels found.")
```

## duplicate in folders

```python
import json
import os
from collections import Counter
```

```python
def find_duplicate_labels_in_folder(folder_path):
    results = {}
    
    for filename in os.listdir(folder_path):
        if filename.endswith(".json"):
            file_path = os.path.join(folder_path, filename)
            
            with open(file_path, 'r', encoding='utf-8') as file:
                data = json.load(file)
            
            labels = []
            
            def extract_labels(obj):
                if isinstance(obj, dict):
                    for key, value in obj.items():
                        labels.append(key)
                        extract_labels(value)
                elif isinstance(obj, list):
                    for item in obj:
                        extract_labels(item)
                elif isinstance(obj, str):
                    labels.append(obj)
            
            extract_labels(data)
            
            label_counts = Counter(labels)
            duplicates = {label: count for label, count in label_counts.items() if count > 1}
            
            if duplicates:
                results[filename] = duplicates
    
    return results

if __name__ == "__main__":
    folder_path = r"C:\Users\tushar.potdukhe\Desktop\TW data\duplicates_labels\duplicates_labels\scooter"  # Change this to your folder containing JSON files
    duplicate_results = find_duplicate_labels_in_folder(folder_path)
    
    if duplicate_results:
        print("Duplicate labels found:")
        for file, duplicates in duplicate_results.items():
            print(f"In {file}:")
            for label, count in duplicates.items():
                print(f"  {label}: {count} times")
    else:
        print("No duplicate labels found in any JSON files.")
```
