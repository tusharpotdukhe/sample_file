```python
import os
import shutil
```

```python
# Define the source directory where the 30 folders are located
source_directory = r"C:\Users\tushar.potdukhe\Desktop\QA_Data_1"

# Define the destination directory where all images will be extracted
destination_directory = r"C:\Users\tushar.potdukhe\Desktop\QAData1"

# Ensure the destination directory exists, create it if not
os.makedirs(destination_directory, exist_ok=True)

# Loop through all folders inside the source directory
for folder in os.listdir(source_directory):
    folder_path = os.path.join(source_directory, folder)
    
    # Check if it's a directory
    if os.path.isdir(folder_path):
        for file in os.listdir(folder_path):
            file_path = os.path.join(folder_path, file)
            
            # Check if it's a file and has an image extension
            if os.path.isfile(file_path) and file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp', '.tiff','.json')):
                
                # Handle duplicate filenames by renaming
                destination_file_path = os.path.join(destination_directory, file)
                counter = 1
                
                while os.path.exists(destination_file_path):
                    name, ext = os.path.splitext(file)
                    new_name = f"{name}_{counter}{ext}"
                    destination_file_path = os.path.join(destination_directory, new_name)
                    counter += 1
                
                # Move the file
                shutil.move(file_path, destination_file_path)
                print(f"Moved: {file_path} -> {destination_file_path}")

print("All images have been extracted successfully!")
```
